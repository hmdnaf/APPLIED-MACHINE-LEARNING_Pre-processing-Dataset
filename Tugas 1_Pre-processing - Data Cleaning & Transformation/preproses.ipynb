```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load dataset baru (separator harus ;)
df = pd.read_csv('dataset_pinjaman_online_converted.csv', sep=';')

print("5 Data Awal:")
df.head()
```

```python
print(df.info())
print(df.describe(include='all'))
```

```python
print("Missing value tiap kolom:")
print(df.isna().sum())
```

```python
# Kolom numerik → isi dengan median
num_cols = ['Umur', 'Pendapatan', 'JumlahPinjaman', 'LamaBekerja']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Kolom kategori → isi mode
cat_cols = ['Pendidikan', 'StatusPerkawinan', 'Pekerjaan', 
            'TujuanPinjaman', 'KodeKota', 'RiwayatPinjaman']
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
le = LabelEncoder()

encode_cols = [
    'Pendidikan',
    'StatusPerkawinan',
    'Pekerjaan',
    'TujuanPinjaman',
    'KodeKota',
    'RiwayatPinjaman',
    'StatusPinjaman'
]

for col in encode_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df_clean = df.drop(columns=['ID_Pelanggan'])
df_clean.head()
```

```python
X = df_clean.drop(columns=['StatusPinjaman'])
y = df_clean['StatusPinjaman']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Ukuran Train:", X_train.shape)
print("Ukuran Test :", X_test.shape)
```

```python
train_df = X_train.copy()
train_df['StatusPinjaman'] = y_train

test_df = X_test.copy()
test_df['StatusPinjaman'] = y_test

train_df.to_csv("data_train_preprocessed.csv", index=False)
test_df.to_csv("data_test_preprocessed.csv", index=False)

print("Dataset berhasil disimpan:")
print("- data_train_preprocessed.csv")
print("- data_test_preprocessed.csv")
```
